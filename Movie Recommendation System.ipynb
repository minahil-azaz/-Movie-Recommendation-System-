```python
import pandas as pd 
import numpy as np 
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
df.shape
```

```python
select_features = ['genres','keywords','tagline','cast','director']
select_features
```

```python
for features in select_features:
    df[features] = df[features].fillna('')
```

```python
df['combined_features'] = (
    df['genres'] + " " +
    df['keywords'] + " " +
    df['tagline'] + " " +
    df['cast'] + " " +
    df['director']
)
```

```python
df['combined_features']
```

```python
df['combined_features'] = df['combined_features'].fillna('')
```

```python
vectorize = TfidfVectorizer(stop_words='english')
```

```python
feature_vector = vectorize.fit_transform(df['combined_features'])
```

```python
feature_vector
```

```python
similarity = cosine_similarity(feature_vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
movies_name = input('enter movie name')
```

```python
list_of_all_movie_name = df['title'].tolist()
```

```python
list_of_all_movie_name
```

```python
find_close_match = difflib.get_close_matches(movies_name,list_of_all_movie_name)
find_close_match
```

```python
close_match = find_close_match[0]
close_match
```

```python
index_of_movie = df[df.title == close_match]['index'].values[0]
index_of_movie
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
similarity_score
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)
sorted_similar_movies
```

```python
print('here are the movies i suggest u')
i =1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if i<30:
        print(i," ", title_from_index)
        i +=1
```

```python
movies_name = input('enter movie name')

list_of_all_movie_name = df['title'].tolist()

find_close_match = difflib.get_close_matches(movies_name,list_of_all_movie_name)
close_match = find_close_match[0]

index_of_movie = df[df.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse=True)

print('here are the movies i suggest u')
i =1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if i<30:
        print(i," ", title_from_index)
        i +=1

```
